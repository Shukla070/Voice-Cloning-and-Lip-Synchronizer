```python
def convert_mp3_to_wav(mp3_path,wav_path,target_sr):
  try:
    audio=AudioSegment.from_mp3(mp3_path)
    audio=audio.set_channels(1).set_frame_rate(target_sr)
    audio.export(wav_path,format="wav")
    print(f"Converted MP3 to WAV: {wav_path}")
    return wav_path
  except Exception as e:
    print(f"Error converting {mp3_path}: {e}")
    return None
```

```python
input_file="/kaggle/input/trumpvoice/Trump/Trump_WEF_2018.mp3"
if input_file.endswith(".mp3"):
  input_wav="/kaggle/working/Trump_WEF_2018"+".wav"
  input_wav=convert_mp3_to_wav(input_file,input_wav,22050)
else:
  input_wav=input_file
```

```python
def split_audio(file_path,output_base,target_sr,duration,max,bit_depth, name):
  try:
    audio,sr=librosa.load(file_path,sr=target_sr,mono=True)
    total_duration=len(audio)/sr
    clip_samples=int(duration*sr)
    interval=total_duration/max
    start_times=[int(i*interval*sr) for i in range(max)]
    for i,start_sample in enumerate(start_times):
      end_sample=start_sample+clip_samples
      clip=audio[start_sample:end_sample]
      output_path=os.path.join(f"/kaggle/working/{name}_voices",f"{output_base}_part{i+1}.wav")
      sf.write(output_path, clip, target_sr,subtype=bit_depth)
      print(f"Saved: {output_path}")
  except Exception as e:
    print(f"Error processing {file_path}: {e}")
```

```python
transcriptions=''
with open("/kaggle/input/archive/archive/transcript.txt","r",encoding='utf-8',errors="ignore") as file:
    file_content=file.read()
    lines=file_content.split('\n')
    transcripts=[line.split('|')[1] for line in lines if '|' in line]
    for line in file.readlines():
        transcriptions+=' '+ line.split('|')[1]
    audio_files=[line.split('|')[0] for line in lines if '|' in line]
```

```python
import re
import torch
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.trainers import BpeTrainer
from german_transliterate.core import GermanTransliterate
def text_cleaners(text):
  text=GermanTransliterate().transliterate(text)
  text=text.lower()
  text=re.sub(re.compile(r'\s+'), ' ', text)
  text=text.replace('"', '')
  return text

def remove_extraneous_punctuation(word):
    replacement_punctuation = {
        '{': '(', '}': ')',
        '[': '(', ']': ')',
        '`': '\'', '—': '-',
        '—': '-', '`': '\'',
        'ʼ': '\''
    }
    replace=re.compile("|".join([re.escape(k) for k in sorted(replacement_punctuation, key=len, reverse=True)]), flags=re.DOTALL)
    word=replace.sub(lambda x: replacement_punctuation[x.group(0)], word)
    extraneous=re.compile(r'^[@#%_=\$\^&\*\+\\]$')
    word=extraneous.sub('', word)
    return word

with open('transcriptions.txt', 'r', encoding='utf-8') as at:
    ttsd = at.readlines()
    allowed_characters_re = re.compile(r'^[a-zäöüß!:;"/, \-\(\)\.\'\?ʼ]+$')

    def preprocess_word(word, report=False):
        word = text_cleaners(word)
        word = remove_extraneous_punctuation(word)
        if not bool(allowed_characters_re.match(word)):
            if report and word:
                print(f"REPORTING: '{word}'")
            return ''
        return word

    def batch_iterator(batch_size=1000):
        print("Processing ASR texts.")
        for i in range(0, len(ttsd), batch_size):
            yield [preprocess_word(t, True) for t in ttsd[i:i+batch_size]]

    trainer = BpeTrainer(special_tokens=['[STOP]', '[UNK]', '[SPACE]'], vocab_size=255)
    tokenizer = Tokenizer(BPE(unk_token="[UNK]"))
    tokenizer.pre_tokenizer = Whitespace()
    tokenizer.train_from_iterator(batch_iterator(), trainer, length=len(ttsd))
    tokenizer.save('../custom_language_tokenizer.json')
```

```python
import librosa
import soundfile as sf
import os
def resample_wav_file(input_file, target_sampling_rate=22050):
    audio, sampling_rate = librosa.load(input_file, sr=None)
    if sampling_rate != target_sampling_rate:
        audio_resampled = librosa.resample(audio, orig_sr=sampling_rate, target_sr=target_sampling_rate)
        sf.write(input_file, audio_resampled, target_sampling_rate)
dataset_path = "/kaggle/input/archive/archive/achtgesichterambiwasse"
for wav_file in os.listdir(dataset_path):
    input_file=os.path.join(dataset_path,wav_file)
    resample_wav_file(input_file)
```

```python
base_audio_path="/kaggle/working/dataset/wavs/"
transcribed_audio_samples = []
for filename, transcript in zip(audio_files,transcripts):
    filename=filename.split("/")[1]
    transcribed_audio_samples.append((f"{base_audio_path}{filename}", f"{filename}|{transcript}"))
```

```python
import random
random.shuffle(transcribed_audio_samples)
num_train_samples = int(len(transcribed_audio_samples) * 0.85)
train_dataset = transcribed_audio_samples[:num_train_samples]
val_dataset = transcribed_audio_samples[num_train_samples:]
```

```python
import shutil

train_val_list = [('train', train_dataset), ('val', val_dataset)]

for stage, dataset in train_val_list:
    sample_list = []
    for old_sample, new_sample in dataset:
        sample_list.append(new_sample)

    with open(f"/kaggle/working/dataset/{stage}.txt", "w",encoding="utf-8",errors="ignore") as sample_file:
        sample_file.write("\n".join(sample_list))
```
